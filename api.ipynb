```python
import requests
import json

# Define the API endpoint and your API key
url = "https://api.elsevier.com/content/search/scopus"
api_key = "35c0b3b495c330e4a0867a611fbcb22b"

# Define the query parameters
params = {
    "query": "PUBYEAR = 2024",
    "apiKey": api_key
}

# Make the request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    try:
        data = response.json()
        # Pretty print the JSON data
        #print(json.dumps(data, indent=4))
        # Save the JSON data to a file
        with open('scopus_data.json', 'w') as json_file:
            json.dump(data, json_file, indent=4)
    except json.JSONDecodeError:
        print("Error: Response content is not valid JSON")
else:
    print(f"Error: {response.status_code}", response.text)

# Define the new API endpoint for abstract retrieval
abstract_url = "https://api.elsevier.com/content/abstract/scopus_id/85208689386"
abstract_params = {
    "apiKey": api_key,
    "httpAccept": "application/json",
    "view" : "META"
}

# Make the request for the abstract
abstract_response = requests.get(abstract_url, params=abstract_params)

# Check if the request was successful
if abstract_response.status_code == 200:
    print(abstract_response.text)
else:
    print(f"Error: {abstract_response.status_code}", abstract_response.text)
```

```python
auth_url = "https://api.elsevier.com/authenticate?platform=SCOPUS"
auth_params = {
    "apiKey": api_key
}

# Make the request for authentication
auth_response = requests.get(auth_url, params=auth_params)

# Check if the request was successful
if auth_response.status_code == 200:
    print(auth_response.text)
else:
    print(f"Error: {auth_response.status_code}", auth_response.text)
```
